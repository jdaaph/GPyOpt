```python
%pylab inline
%load_ext autoreload
%autoreload 2

from GPyOpt.methods import BayesianOptimization
import GPy
import GPyOpt
from numpy.random import seed
from tqdm import *
import io
# import matplotlib.pyplot as plt
```

```python
def myf(x):
    return -((x>0)*2-1)

bounds = [{'name': 'var_1', 'type': 'continuous', 'domain': (-100,100)}]
max_iter = 15
```

```python
myProblem = BayesianOptimization(myf,bounds,acquisition_type="BinaryT",
#                                  initial_design_numdata=None,
                                 normalize_Y = True,
                                 batch_size = 1,
                                 evaluator_type='local_penalization')
```

```python
from IPython.display import clear_output
# myProblem.plot_acquisition()

for i in tqdm(range(6)):
    myProblem.run_optimization(max_iter=1) 
    fig = myProblem.plot_acquisition_binaryT()
```
