```python
%pylab inline  
import GPyOpt
```

```python
f_true= GPyOpt.objective_examples.experiments1d.forrester()          # noisy version
bounds = [{'name': 'var_1', 'type': 'continuous', 'domain': (0,1)}]  # problem constrains 
f_true.plot()
```

```python
batch_size = 5
num_cores = 1
```

```python
from numpy.random import seed
seed(123)
BO_demo_parallel = GPyOpt.methods.BayesianOptimization(f=f_true.f,  
                                            domain = bounds,                  
                                            acquisition_type = 'BinaryT',              
                                            normalize_Y = True,
                                            initial_design_numdata = 5,
                                            evaluator_type = 'local_penalization',
                                            batch_size = batch_size,
                                            num_cores = num_cores,
                                            acquisition_jitter = 0)  
```

```python
# --- Run the optimization for 10 iterations
max_iter = 10 
for i in range(max_iter):
    BO_demo_parallel.plot_acquisition()
    BO_demo_parallel.run_optimization(1)
```
